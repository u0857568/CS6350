```python
def toNum(data):
    data['workclass'] = data['workclass'].astype(str)
    data['workclass'].replace({'?':0, 'Self-emp-not-inc':1,'Private':2,'Federal-gov':3,'Local-gov':4,'Self-emp-inc':5,'State-gov':6,'Never-worked':7,'Without-pay':8}, inplace=True)
    data['education'] = data['education'].astype(str)
    data['education'].replace({'Prof-school':0, 'Bachelors':1,'HS-grad':2,'Doctorate':3,'Masters':4,'Some-college':5,'11th':6,'12th':7,'5th-6th':8,'Assoc-acdm':9,'9th':10,'Preschool':11,'Assoc-voc':12,'7th-8th':13,'10th':14,'1st-4th':15}, inplace=True)
    data['marital.status'] = data['marital.status'].astype(str)
    data['marital.status'].replace({'Married-civ-spouse':0, 'Divorced':1,'Never-married':2,'Married-spouse-absent':3,'Widowed':4,'Separated':5,'Married-AF-spouse':6}, inplace=True)
    data['occupation'] = data['occupation'].astype(str)
    data['occupation'].replace({'Prof-specialty':0, 'Exec-managerial':1,'Craft-repair':2,'Transport-moving':3,'Other-service':4,'Machine-op-inspct':5,'Adm-clerical':6,'Sales':7,'Handlers-cleaners':8,'?':9,'Farming-fishing':10,'Tech-support':11,'Priv-house-serv':12,'Protective-serv':13,'Armed-Forces':14}, inplace=True)
    data['relationship'] = data['relationship'].astype(str)
    data['relationship'].replace({'Husband':0, 'Other-relative':1,'Unmarried':2,'Not-in-family':3,'Own-child':4,'Wife':5}, inplace=True)
    data['race'] = data['race'].astype(str)
    data['race'].replace({'Asian-Pac-Islander':0, 'White':1,'Black':2,'Other':3,'Amer-Indian-Eskimo':4}, inplace=True)
    data['sex'] = data['sex'].astype(str)
    data['sex'].replace({'Male':0, 'Female':1}, inplace=True)
    data['native.country'] = data['native.country'].astype(str)
    data['native.country'].replace({'India':0, 'United-States':1,'Vietnam':2,'England':3,'Mexico':4,'?':5,'Guatemala':6,'Iran':7,'Laos':8,'Germany':9,'Thailand':10,'Ireland':11,'Philippines':12,'Hong':13,'China':14,'Dominican-Republic':15,'Scotland':16,'Haiti':17,'Cuba':18,'Jamaica':19,'Peru':20,'Canada':21,'Cambodia':22,'Italy':23,'Poland':24,'Yugoslavia':25,'El-Salvador':26,'Columbia':27,'Greece':28,'Ecuador':29,'Japan':30,'South':31,'Puerto-Rico':32,'Nicaragua':33,'Outlying-US(Guam-USVI-etc)':34,'Taiwan':35,'Honduras':36,'Trinadad&Tobago':37,'Portugal':38,'Hungary':39,'France':40,'Holand-Netherlands':41}, inplace=True)
    return data
    
```

```python
y = train['income>50K']
x = pd.get_dummies(train.drop('income>50K', axis=1))

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)
```

```python
def fitAndPredict(model):
    
    model.fit(x_train, y_train)
    prediction = model.predict(x_test)
    return accuracy_score(y_test, prediction)

```

```python
model1 = LogisticRegression(solver='liblinear', random_state=42)
model2 = GradientBoostingClassifier()
model3 = RandomForestClassifier()
model4 = SGDClassifier()
model5 = SVC()
model6 = MLPClassifier()

models = [model1, model2, model3, model4, model5, model6]
i = 0
for model in models:
    i +=1
    print("Model ", i,":", model)
    print("ACC: ", fitAndPredict(model))
```

```python
model = GradientBoostingClassifier(min_samples_split=20, min_samples_leaf=60, max_depth=3, max_features=7)
fitAndPredict(model)
```
